### Series of annual maxima

```python
annualMax = pd.DataFrame(index=pcp_h.index.year.unique())
for d in np.arange(1, 25):
    serie = pcp_h.rolling(d, center=True).sum()
    annualMax[d] = serie.groupby(serie.index.year).max() / d
annualMax.dropna(axis=0, how='all', inplace=True)
annualMax
```

```python
from scipy.stats import genextreme
from statsmodels.distributions.empirical_distribution import ECDF
import seaborn as sns
sns.set()
```

```python
sns.distplot(annualMax[12], bins=8);
```

```python
# Example
# fit a GEV to the 1h-duration storm
par_int1h = genextreme.fit(annualMax[1], f0=0)
print('Fitted parameters for the intensity in 1h-duration storms:')
print('xi = {0:.4f}\nmu = {1:.4f}\nsigma = {2:.4f}'.format(*par_int1h))
```

```python
# fit parameters for the 3 durations
parameters = pd.DataFrame(index=['xi', 'mu', 'sigma'], columns=annualMax.columns)
for duration in annualMax.columns:
    # Ajustar la GEV y guardar los parámetros
    parameters[duration] = genextreme.fit(annualMax[duration], f0=0)
parameters
```

```python
P = np.linspace(start=0, stop=30, num=500)
for d in [1, 2, 4, 6, 12, 24]: 
    plt.plot(P, genextreme(*parameters[d]).cdf(P))
    plt.scatter(annualMax[d], ECDF(annualMax[d])(annualMax[d]), s=5)
```

```python
# return periods
R = np.array([10, 50, 100], dtype="float64")
```

```python
# non-exceedance probability
Q = 1. - 1. / R
```

```python
# extreme values of intensity for a 1-h storm
P1 = genextreme.ppf(Q, *parameters[1]) # ppf: percent point function

print('Rainfall intensity in 1 h storms at different return perios:')
for i, Tr in enumerate(R):
     print('P(Tr=', int(Tr), ') = ', round(P1[i], 1), ' mm/h', sep='')
```

```python
# data frame with values of the IDF curve
IDFe = pd.DataFrame(index=R, columns=annualMax.columns)
IDFe.index.name = 'Tr'
for duration in annualMax.columns:
    IDFe[duration] = genextreme(*parameters[duration]).ppf(Q)
IDFe
```

```python
# configuración del gráfico
fig = plt.figure(figsize=(12, 6))
plt.title('Empirical IDF', fontsize=16, weight='bold')
plt.xlabel('duration (h)', fontsize=13)
plt.xlim(0, IDFe.columns.max() + 1)
plt.ylabel('intensity (mm/h)', fontsize=13)
plt.ylim((0, 40))
color = ['tan', 'darkkhaki', 'olive', 'darkolivegreen']

for i, Tr in enumerate(IDFe.index):
    plt.scatter(IDFe.columns, IDFe.loc[Tr,:], color=color[i], label='Tr = ' + str(int(Tr)) + ' years')

fig.legend(loc=8, ncol=4, fontsize=12);

# guardar la figura
#plt.savefig('../output/Ex5_empirical IDFs.png', dpi=300)
```

```python
def IDF_type_0(x, a, b):
    """Estimate precipitation intensity fiven a return period and a storm duration using the analytical IDF curve type IV:
    
    I = a * x**2 / (x**2 + b).    
    
    Input:
    ------
    x:         list [2x1]. Values of return period (years) and duration (h)
    a:         float. Parameter of the IDF curve
    b:         float. Parameter of the IDF curve
    
    Output:
    -------
    I:         float. Precipitation intensity (mm/h)"""
    
    I = a * x**2 / (x**2 + b)
    return I 
```

```python
def IDF_type_I(x, b, c, d, e):
    """Estimate precipitation intensity fiven a return period and a storm duration using the analytical IDF curve type I:
    
    I = (d * R + e) / (D + c)**b.    
    
    Input:
    ------
    x:         list [2x1]. Values of return period (years) and duration (h)
    b:         float. Parameter of the IDF curve
    c:         float. Parameter of the IDF curve
    d:         float. Parameter of the IDF curve
    e:         float. Parameter of the IDF curve
    
    Output:
    -------
    I:         float. Precipitation intensity (mm/h)"""
    
    a = d * x[0] + e
    I = a / (x[1] + c)**b
    
    return I
```

```python
from scipy.optimize import curve_fit
```

```python
Tr = 100
```

```python
D = np.arange(1, 25)
R = np.ones_like(D) * Tr
RD = np.vstack([R, D])
```

```python
# fit the curve
curve = IDF_type_I
popt, pcov = curve_fit(curve, RD, IDFe.loc[Tr,:])

print('Fitted parameters of the analytical IDF')
for i, par in enumerate(['b', 'c', 'd', 'e']):
    print(par, '=', round(popt[i], 4))
```

```python
D_ = np.linspace(1, 25, 100)
R_ = np.ones_like(D_) * Tr
RD_ = np.vstack((R_, D_))
y_ = IDF_type_I(RD_, *popt)
```
